```python
import pandas
import matplotlib.pyplot as pyplot
```

```python
# read clean data
units_df = pandas.read_csv('units_cleaned.csv')
```

```python
# get only mechs and omnimechs
mechs_df = units_df[units_df['Type'].isin(['BattleMech ', 'BattleMech - OmniMech'])]
print(len(mechs_df))

battle_mechs_df = units_df[units_df['Type'] == 'BattleMech ']
print(len(battle_mechs_df))

omni_mechs_df = units_df[units_df['Type'] == 'BattleMech - OmniMech']
print(len(omni_mechs_df))
```

```python
# compute cost efficiency
mechs_df[['Cost', 'BattleValue', 'Tonnage']] = mechs_df[['Cost', 'BattleValue', 'Tonnage']].apply(pandas.to_numeric)
mechs_df['Ratio']                            = mechs_df['Cost'] / mechs_df['BattleValue']

battle_mechs_df[['Cost', 'BattleValue', 'Tonnage']] = battle_mechs_df[['Cost', 'BattleValue', 'Tonnage']].apply(pandas.to_numeric)
battle_mechs_df['Ratio']                            = battle_mechs_df['Cost'] / battle_mechs_df['BattleValue']

omni_mechs_df[['Cost', 'BattleValue', 'Tonnage']] = omni_mechs_df[['Cost', 'BattleValue', 'Tonnage']].apply(pandas.to_numeric)
omni_mechs_df['Ratio']                            = omni_mechs_df['Cost'] / omni_mechs_df['BattleValue']
```

```python
# range brackets
light_range   = [20, 25, 30, 35]
medium_range  = [40, 45, 50, 55]
heavy_range   = [60, 65, 70, 75]
assault_range = [80, 85, 90, 95, 100]
```

```python
# plot all data
mechs_df.plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='All Mechs')

# plot mech data by weight class
mechs_df[mechs_df['Tonnage'].isin(light_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Light Mechs')
mechs_df[mechs_df['Tonnage'].isin(medium_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Medium Mechs')
mechs_df[mechs_df['Tonnage'].isin(heavy_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Heavy Mechs')
mechs_df[mechs_df['Tonnage'].isin(assault_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Assault Mechs')
```

```python
# plot all data
battle_mechs_df.plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='All BattleMechs')

# plot mech data by weight class
battle_mechs_df[battle_mechs_df['Tonnage'].isin(light_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Light BattleMechs')
battle_mechs_df[battle_mechs_df['Tonnage'].isin(medium_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Medium BattleMechs')
battle_mechs_df[battle_mechs_df['Tonnage'].isin(heavy_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Heavy BattleMechs')
battle_mechs_df[battle_mechs_df['Tonnage'].isin(assault_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Assault BattleMechs')
```

```python
# plot all data
omni_mechs_df.plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='All OmniMechs')

# plot mech data by weight class
omni_mechs_df[omni_mechs_df['Tonnage'].isin(light_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Light OmniMechs')
omni_mechs_df[omni_mechs_df['Tonnage'].isin(medium_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Medium OmniMechs')
omni_mechs_df[omni_mechs_df['Tonnage'].isin(heavy_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Heavy OmniMechs')
omni_mechs_df[omni_mechs_df['Tonnage'].isin(assault_range)].plot.hist(bins=30, column=['Ratio'], range=(0, 20000), title='Assault OmniMechs')
```
