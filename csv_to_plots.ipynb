```python
import pandas
import matplotlib.pyplot as pyplot
```

```python
# read clean data
units_df = pandas.read_csv('units_cleaned.csv')
```

```python
# get only mechs and omnimechs
mechs_df = units_df[units_df['Type'].isin(['BattleMech ', 'BattleMech - OmniMech'])]
print(len(mechs_df))

battle_mechs_df = units_df[units_df['Type'] == 'BattleMech ']
print(len(battle_mechs_df))

omni_mechs_df = units_df[units_df['Type'] == 'BattleMech - OmniMech']
print(len(omni_mechs_df))
```

```python
# compute cost efficiency
mechs_df[['Cost', 'BattleValue', 'Tonnage']] = mechs_df[['Cost', 'BattleValue', 'Tonnage']].apply(pandas.to_numeric)
mechs_df['Ratio']                            = mechs_df['Cost'] / mechs_df['BattleValue']

battle_mechs_df[['Cost', 'BattleValue', 'Tonnage']] = battle_mechs_df[['Cost', 'BattleValue', 'Tonnage']].apply(pandas.to_numeric)
battle_mechs_df['Ratio']                            = battle_mechs_df['Cost'] / battle_mechs_df['BattleValue']

omni_mechs_df[['Cost', 'BattleValue', 'Tonnage']] = omni_mechs_df[['Cost', 'BattleValue', 'Tonnage']].apply(pandas.to_numeric)
omni_mechs_df['Ratio']                            = omni_mechs_df['Cost'] / omni_mechs_df['BattleValue']
```

```python
# range brackets
light_range   = [20, 25, 30, 35]
medium_range  = [40, 45, 50, 55]
heavy_range   = [60, 65, 70, 75]
assault_range = [80, 85, 90, 95, 100]
```

```python
# plot all data
_ = mechs_df.plot.hist(
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='All Mechs'
)

# plot mech data by weight class
fig, axes = pyplot.subplots(2, 2, figsize=(10,10))
_ = mechs_df[mechs_df['Tonnage'].isin(light_range)].plot.hist(
    ax=axes[0,0],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Light Mechs'
)
_ = mechs_df[mechs_df['Tonnage'].isin(medium_range)].plot.hist(
    ax=axes[0,1],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Medium Mechs'
)
_ = mechs_df[mechs_df['Tonnage'].isin(heavy_range)].plot.hist(
    ax=axes[1,0],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Heavy Mechs'
)
_ = mechs_df[mechs_df['Tonnage'].isin(assault_range)].plot.hist(
    ax=axes[1,1],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Assault Mechs'
)
```

```python
# plot all data
_ = battle_mechs_df.plot.hist(
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='All BattleMechs'
)

# plot mech data by weight class
fig, axes = pyplot.subplots(2, 2, figsize=(10,10))
_ = battle_mechs_df[battle_mechs_df['Tonnage'].isin(light_range)].plot.hist(
    ax=axes[0,0],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Light BattleMechs'
)
_ = battle_mechs_df[battle_mechs_df['Tonnage'].isin(medium_range)].plot.hist(
    ax=axes[0,1],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Medium BattleMechs'
)
_ = battle_mechs_df[battle_mechs_df['Tonnage'].isin(heavy_range)].plot.hist(
    ax=axes[1,0],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Heavy BattleMechs'
)
_ = battle_mechs_df[battle_mechs_df['Tonnage'].isin(assault_range)].plot.hist(
    ax=axes[1,1],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Assault BattleMechs'
)
```

```python
# plot all data
_ = omni_mechs_df.plot.hist(
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='All OmniMechs'
)

# plot mech data by weight class
fig, axes = pyplot.subplots(2, 2, figsize=(10,10))
_ = omni_mechs_df[omni_mechs_df['Tonnage'].isin(light_range)].plot.hist(
    ax=axes[0,0],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Light OmniMechs'
)
_ = omni_mechs_df[omni_mechs_df['Tonnage'].isin(medium_range)].plot.hist(
    ax=axes[0,1],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Medium OmniMechs'
)
_ = omni_mechs_df[omni_mechs_df['Tonnage'].isin(heavy_range)].plot.hist(
    ax=axes[1,0],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Heavy OmniMechs'
)
_ = omni_mechs_df[omni_mechs_df['Tonnage'].isin(assault_range)].plot.hist(
    ax=axes[1,1],
    bins=30,
    column=['Ratio'],
    range=(0, 20000),
    title='Assault OmniMechs'
)
```
